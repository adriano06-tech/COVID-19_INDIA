# Conhecendo os dados da covid

## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize': (12, 6)})
print('Todas as bibliotecas foram importadas com sucesso')
```

## Preparando o ambiente

```python
PRINCIPAL_PATH = './data/'
SEED = 101
np.random.seed(SEED)
```

## Lendo a base de dados

```python
covid_india = pd.read_csv(PRINCIPAL_PATH + 'covid_india/raw_covid_19_india.csv')
covid_india.head()
```

## Tratando os dados

```python
covid_india.rename(
    columns = {
        'Cured': 'CumulativeCured',
        'Deaths': 'CumulativeDeaths',
        'Confirmed': 'CumulativeConfirmed'
    },
    inplace = True
)

covid_india.head()
```

```python
covid_india.isna().sum()
```

```python
covid_india.dropna(inplace = True)
covid_india.isna().sum()
```

```python
covid_india.query('ConfirmedIndianNational == "-"')
```

```python
covid_india.drop(columns = ['ConfirmedIndianNational', 'ConfirmedForeignNational'], inplace = True)
covid_india.head()
```

```python
covid_india.drop(columns = ['Sno'], inplace = True)
covid_india.head()
```

```python
covid_india.head()
```

```python
cumulative_variables = ['CumulativeCured', 'CumulativeDeaths', 'CumulativeConfirmed']
for var in cumulative_variables:
    covid_india[var] = covid_india[var].astype(int)
covid_india.head()
```

```python
covid_india['Date'] = pd.to_datetime(covid_india['Date'])
covid_india.dtypes
```

```python
covid_india['Time'].drop_duplicates()
```

```python
time_to_24_rename = {}

for time in covid_india['Time'].drop_duplicates():
    time_type = time[-2:]
    time_numerics = time[:-3]
    hour = int(time_numerics.split(':')[0])
    minute = time_numerics.split(':')[-1]
    if time_type == 'PM':
        hour += 12
    time_to_24_rename[time] = str(hour) + ':' + str(minute)

time_to_24_rename
```

```python
covid_india['Time'] = covid_india['Time'].map(time_to_24_rename)
covid_india.head()
```

```python
new_datetime = [
    pd.Timestamp(str(row['Date']) + ' ' + row['Time'])
    for index, row in covid_india.iterrows()
]

covid_india['DateTime'] = new_datetime
covid_india.head()
```

```python
covid_india.drop(columns = ['Time'], inplace = True)
covid_india.head()
```

```python
covid_india.rename(columns = {'State/UnionTerritory': 'Territory'}, inplace = True)
covid_india.head()
```

```python
covid_india['CumulativeDeaths'].sum()
```

```python
cumulatives_variables_by_territory = covid_india.groupby(['Territory', 'Date']).sum()
cumulatives_variables_by_territory.head()
```

```python
new_covid_india = {'Date': [], 'Territory': []}

for var in cumulative_variables:
    non_cumulative_var_name = var.replace('Cumulative', '')
    new_covid_india[non_cumulative_var_name] = []


for territory in covid_india['Territory'].drop_duplicates():
    territory_cumulatives = cumulatives_variables_by_territory.loc[territory]
    for index, row in territory_cumulatives.iterrows():
        new_covid_india['Date'].append(index)
        new_covid_india['Territory'].append(territory)
    
    for var in cumulative_variables:
        non_cumulative_var_name = var.replace('Cumulative', '')
        new_covid_india[non_cumulative_var_name] += list(territory_cumulatives[var].diff())
        
        
new_covid_india = pd.DataFrame(new_covid_india)
new_covid_india.fillna(0, inplace = True)
new_covid_india.head()
```

```python
covid_india = new_covid_india
covid_india.head()
```

```python
covid_india.sum()
```

```python
covid_india.dtypes
```

```python
numeric_variables = ['Cured', 'Deaths', 'Confirmed']
for var in numeric_variables:
    covid_india[var] = covid_india[var].astype(int)
covid_india.head()
```

```python
covid_india = covid_india.query('Cured > 0 and Deaths > 0 and Confirmed > 0')
covid_india.describe()
```

## Salvando todas as alterações feitas

```python
covid_india.to_csv(PRINCIPAL_PATH + 'covid_india/new_covid_19_india.csv', index = False)
print('Salvo com sucesso!')
```
